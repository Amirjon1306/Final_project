# Загружаем необходимые библиотеки и считываем наши данные

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report

data = pd.read_csv('diabetes_prediction_dataset.csv.zip')
data
```

# Проверяем на пропуски значений

```python
a=data.value_counts('smoking_history')
a
```

# Удаляем строки с неизвестными данными

```python
data = data[data['smoking_history'] != 'No Info']
data
```

# Создаем матрицу корреляций для выявления взаимосвязи между переменными

```python
label_encoder = LabelEncoder()
for col in data.select_dtypes(include=['object']).columns:
    data[col] = label_encoder.fit_transform(data[col])

corr_matrix = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()
```

# Создаём модель и обучаем её

```python
# Делим данные на признаки и целевые переменные
X = data.drop('diabetes', axis=1)
y = data['diabetes']
# Стандаритзируем данные
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Разделение данных на обучающие и тестовые 
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
# Создание модели дерево решений
model = DecisionTreeClassifier(random_state=42) 
model.fit(X_train, y_train)
# Предсказываем модель и используем матрицу ошибок для выявление эффективности модели
y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
# Создаем гистограмму чтобы понять какие признаки важны для нашей модели
feature_importances = pd.Series(model.feature_importances_, index=data.drop('diabetes', axis=1).columns)
feature_importances.sort_values(ascending=False).plot(kind='bar', figsize=(12, 6))
plt.title('Важность признаков')
plt.show()
```

# Строим и визуализируем кривую ROC для оценки производительности модели

```python
from sklearn.metrics import roc_auc_score, roc_curve, auc
import matplotlib.pyplot as plt
y_pred = model.predict(X_test)
y_pred_prob = model.predict_proba(X_test)[:, 1]

roc_auc = roc_auc_score(y_test, y_pred_prob)
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
roc_auc_value = auc(fpr, tpr)

print(f"AUC-ROC: {roc_auc}")

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (area = {roc_auc_value:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

# Модели которые я использовал для выявления самой эффективной для моего проекта

```python
# Модель RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Модель LogisticRegression
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=42) 
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#Модель GradientBoostingClassifier
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier(
    max_depth=12,
    min_samples_leaf=7,
    min_samples_split=3,
    n_estimators=486,
    random_state=42
)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Модель КИИ
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5) 
model.fit(X_train, y_train) 
y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Модель DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=42) 
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
